# Data Collection Example Notebook

The aim of this notebook is to show the methodology used to download all lightcurves available from the Kepler Mission. This method combines the lightcurve package (as demonstrated in the previous notebook) to download the curves themselves, with the list of KIC's being provided by the csv file (part of which is shown previously). 

### 1. Installing and Import Lightkurve

```
!pip install lightkurve
import lightkurve as lk
```

### 2. Reading the csv file of KIC's

```
import pandas as pd
```

This file may seem different than the above 'Kepler Cumulative Table', however it is from the same source but only the necessary columns were kept.

```
df = pd.read_csv('Kepler_Data_Raw.csv',skiprows=9)
df
```

```
for kepid in df['kepid'].iloc[0:5]:
  try: 

    # FORMATTING THE KEPID STRING AND DOWNLOADING THE DATA FROM LIGHTKURVE
    KIC = 'KIC ' + str(kepid)
    lc = lk.search_lightcurve(KIC, author='kepler', quarter = 4, cadence='long').download()
    lc.plot()


  except: 
    pass
```
