### 2. Searching and Downloading Lightcurves

To download lightcurves for specific Kepler Input Catalog (KIC) numbers, the syntax below can be used. This returns a table with information about the data available for the KIC. In this particular case, KIC 3733346 was observed in 18 different quarters 

```
search_result = lk.search_lightcurve('KIC 3733346', author='Kepler')
search_result
```

### 3. Plotting a lightcurve

To plot a lightcurve, we can select a specific quarter, download it, then plot it with lightkurve's in-built plotting method. 

```
search_result_q2 = lk.search_lightcurve('KIC 3733346', author='Kepler', quarter=2)
lc = search_result_q2.download()
lc.plot()
```
